```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/OLX_cars_dataset00.csv")
df.head()
```

```python
plt.figure(figsize=(8, 5))
df["Price"].plot(kind="hist", bins=50, color="steelblue", edgecolor="black")
plt.title("Price Distribution")
plt.xlabel("Price (PKR)")
plt.tight_layout()
plt.savefig("/content/plot1_price_distribution.png", dpi=150)
plt.show()
```

```python
brand_price = df.groupby("Make")["Price"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
brand_price.plot(kind="bar", color="coral", edgecolor="black")
plt.title("Brand vs Mean Price")
plt.ylabel("Mean Price (PKR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("/content/plot2_brand_vs_price.png", dpi=150)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["KM's driven"], df["Price"], alpha=0.2, color="green")
plt.title("Mileage vs Price")
plt.xlabel("KM's Driven")
plt.ylabel("Price (PKR)")
plt.tight_layout()
plt.savefig("/content/plot3_mileage_vs_price.png", dpi=150)
plt.show()
```

```python
city_price = df.groupby("Registration city")["Price"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
city_price.plot(kind="bar", color="mediumpurple", edgecolor="black")
plt.title("City vs Mean Price")
plt.ylabel("Mean Price (PKR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("/content/plot4_city_vs_price.png", dpi=150)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

axes[0,0].imshow(mpimg.imread("/content/plot1_price_distribution.png"))
axes[0,0].axis("off")
axes[0,0].set_title("1. Price Distribution")

axes[0,1].imshow(mpimg.imread("/content/plot2_brand_vs_price.png"))
axes[0,1].axis("off")
axes[0,1].set_title("2. Brand vs Mean Price")

axes[1,0].imshow(mpimg.imread("/content/plot3_mileage_vs_price.png"))
axes[1,0].axis("off")
axes[1,0].set_title("3. Mileage vs Price")

axes[1,1].imshow(mpimg.imread("/content/plot4_city_vs_price.png"))
axes[1,1].axis("off")
axes[1,1].set_title("4. City vs Mean Price")

plt.suptitle("Figure 4.2: EDA Visualisations", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.savefig("/content/figure_4_2.png", dpi=150)
plt.show()
```
